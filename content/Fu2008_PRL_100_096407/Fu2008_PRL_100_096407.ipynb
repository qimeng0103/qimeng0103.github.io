# [PRL 2008, Liang Fu and C.L.Kane] [Superconducting Proximity Effect and Majorana Fermions at the Surface of a Topological Insulator](https://doi.org/10.1103/PhysRevLett.100.096407)

 Nambu 基底 $\Psi = ((\psi_{\uparrow}, \psi_{\downarrow}), (\psi_{\downarrow}^{\dagger}, -\psi_{\uparrow}^{\dagger}))^T$ 下，系统的总哈密顿量为：$$\mathcal{H} = -i v \tau^z (\sigma^x \partial_x + \sigma^y \partial_y) - \mu \tau^z + \Delta(y) (\tau^x \cos \phi(y) + \tau^y \sin \phi(y))$$

在 $L \gg W$ 的长结极限下，系统沿 $x$ 方向具有平移对称性。因此，横向动量 $q$ 是一个守恒的量子数，即波函数携带相因子 $\sim e^{iqx}$，$\partial_x\to iq$。为了能够处理任意化学势 $\mu$ 和能量 $E$ 的情况，我们将各区域的纵向分量拟设为 $\sim e^{ik_{y}^{(i)} y}e^{-\kappa^{(i)}_yy}$, $\partial_y\to ik_{y}^{(i)}-\kappa_y^{(i)}$，其中 $i=I,II,III$代表三个区域。

> 区域 I：$y < -W/2$。该区域 $\Delta = \Delta_0$，相位 $\phi = 0$。$$\mathcal{H}_I = v q (\tau^z \sigma^x)+ v (k_{y}^{I}+i\kappa_y^{(I)})(\tau^z \sigma^y) - \mu_S \tau^z + \Delta_0 \tau^x$$为了满足 $y \to -\infty$ 的束缚条件，$\kappa_y^{(I)}<0$。

> 区域 II：$-W/2 < y < W/2$。该区域 $\Delta = 0$。$$\mathcal{H}_{II} = v q (\tau^z \sigma^x) + v (k_{y}^{II}+i\kappa_y^{(II)})(\tau^z \sigma^y) - \mu_N \tau^z$$

> 区域 III：$y > W/2$。此区域相位为 $\phi$。$$\mathcal{H}_{III} = v q (\tau^z \sigma^x) + v (k_{y}^{III}+i\kappa_y^{(III)})(\tau^z \sigma^y) - \mu_S \tau^z + \Delta_0 (\tau^x \cos \phi + \tau^y \sin \phi)$$
为了满足 $y \to +\infty$ 的束缚条件，$\kappa_y^{(III)}>0$。


```python
import sympy as sp
from sympy.physics.quantum import TensorProduct
from IPython.display import display, Math

# 1. 基础矩阵定义
s0 = sp.eye(2)
sx = sp.Matrix([[0, 1], [1, 0]])
sy = sp.Matrix([[0, -sp.I], [sp.I, 0]])
sz = sp.Matrix([[1, 0], [0, -1]])

# 2. 物理符号定义
v, q, phi, Delta0 = sp.symbols('v q phi Delta_0', real=True, positive=True)
mu_S, mu_N, E = sp.symbols('mu_S mu_N E', real=True)
kyI, kappaI = sp.symbols('k_y^{(I)} \kappa_y^{(I)}', real=True)
kyII, kappaII = sp.symbols('k_y^{(II)} \kappa_y^{(II)}', real=True)
kyIII, kappaIII = sp.symbols('k_y^{(III)} \kappa_y^{(III)}', real=True)

def get_H_matrix(phi, delta, ky, kappa, mu):
    tp_zy = sp.Matrix(TensorProduct(sz, sy))
    tp_zx = sp.Matrix(TensorProduct(sz, sx))
    tp_x0 = sp.Matrix(TensorProduct(sx, s0))
    tp_y0 = sp.Matrix(TensorProduct(sy, s0))
    tp_z0 = sp.Matrix(TensorProduct(sz, s0))
    
    term_dx = tp_zx * (v * q)
    term_dy = v * tp_zy * (ky + sp.I * kappa)
    term_mu = tp_z0 * (-mu)
    term_delta = (tp_x0 * sp.cos(phi) + tp_y0 * sp.sin(phi)) * delta

    
    return sp.simplify(term_dx + term_dy + term_mu + term_delta)


# 3. 构建三段区域
H1 = get_H_matrix(0, Delta0, kyI, kappaI, mu_S)
H2 = get_H_matrix(0, 0, kyII, kappaII, mu_N)
H3 = get_H_matrix(phi, Delta0, kyIII, kappaIII, mu_S)

# 4. 展示
display(Math(r"\text{区域 I (S Region) 哈密顿量：}"))
display(H1)
display(Math(r"\text{区域 II (N Region) 哈密顿量：}"))
display(H2)
display(Math(r"\text{区域 III (S Region) 哈密顿量：}"))
display(H3)
```

### $W=0,\mu=0,q\not=0$

> 首先考虑 $W=0,\mu_{S}=\mu_N=0$ 的情况

```python
H_I_eff = sp.simplify((H1 - E * sp.eye(4)).subs({mu_S: 0}))
H_III_eff = sp.simplify((H3 - E * sp.eye(4)).subs({mu_S: 0}))

det_I = sp.factor(H_I_eff.det())
det_III = sp.factor(H_III_eff.det())
display(Math(r"\det(\mathcal{H}_I - E) = " + sp.latex(det_I) + " = 0"))
display(Math(r"\det(\mathcal{H}_{III} - E) = " + sp.latex(det_III) + " = 0"))
```

```python
det_I_at_ky0 = det_I.subs(kyI, 0)
display(Math(r"\text{分析特征方程的虚部：}2v^2 k_y^{(I)} \kappa_y^{(I)} = 0"))
display(Math(r"\text{由于束缚态要求衰减系数 } \kappa_y^{(I)} \neq 0 \text{，数学上必然导出：}k_y^{(I)} = 0"))
sol_kappa_I = sp.solve(det_I_at_ky0, kappaI)
kappa_I = sp.simplify(sol_kappa_I[0])
display(Math(r"\text{将 } k_y^{(I)} = 0 \text{ 代回特征方程，解得解析衰减系数为：}"))
display(Math(rf"\kappa_y^{{(I)}} = {sp.latex(kappa_I)}"))

det_III_at_ky0 = det_III.subs(kyIII, 0)
sol_kappa_III = sp.solve(det_III_at_ky0, kappaIII)
kappa_III = sp.simplify(sol_kappa_III[1])
display(Math(rf"\kappa_y^{{(III)}} = {sp.latex(kappa_III)}"))
```

```python
H_I_final = H_I_eff.subs(kappaI, kappa_I).subs(kyI, 0)
H_III_final = H_III_eff.subs(kappaIII, kappa_III).subs(kyIII, 0)

# 提取零空间（即满足 (H-E)psi = 0 的波函数基底）
ns_I = [sp.simplify(vec) for vec in H_I_final.nullspace()]
ns_III = [sp.simplify(vec) for vec in H_III_final.nullspace()]

display(Math(r"\text{区域 I 基底 } \psi_{I}:"))
for v in ns_I: display(v)

display(Math(r"\text{区域 III 基底 } \psi_{III}:"))
for v in ns_III: display(v)
```

由于在界面 $y=0$ 处满足 $\Psi_I = \Psi_{III}$，我们得到：$$a_1 \psi_{I,1} + a_2 \psi_{I,2} = b_1 \psi_{III,1} + b_2 \psi_{III,2}$$为了求解这些系数，我们将所有项移到等式左边：$$a_1 \psi_{I,1} + a_2 \psi_{I,2} - b_1 \psi_{III,1} - b_2 \psi_{III,2} = 0$$将此式写成矩阵形式：$$\begin{pmatrix} \vdots & \vdots & \vdots & \vdots \\ \psi_{I,1} & \psi_{I,2} & -\psi_{III,1} & -\psi_{III,2} \\ \vdots & \vdots & \vdots & \vdots \end{pmatrix} \begin{pmatrix} a_1 \\ a_2 \\ b_1 \\ b_2 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \\ 0 \\ 0 \end{pmatrix}$$我们定义这个 $4 \times 4$ 的矩阵为 $W$：$$W = [ \psi_{I,1}, \psi_{I,2}, -\psi_{III,1}, -\psi_{III,2} ]$$

存在非零解的条件：$\det(W) = 0$

```python
# 拼接矩阵：将四个 4x1 的列向量组合成 4x4 矩阵
W = sp.Matrix.hstack(ns_I[0], ns_I[1], -ns_III[0], -ns_III[1])

# 计算行列式
det_W = sp.expand(W.det())
display(Math(r"\det(W) = 0 \implies " + sp.latex(det_W) + " = 0"))

eq_sym = sp.expand(det_W * sp.exp(sp.I*phi))
display(Math(r"\text{对称化后 (乘以 } e^{i\phi} \text{): } " + sp.latex(eq_sym) + " = 0"))

eq_trig = eq_sym.rewrite(sp.cos)
display(Math(r"\text{转化为三角函数形式: } " + sp.latex(eq_trig) + " = 0"))

sol_E = sp.solve(eq_trig, E)

for i, sol in enumerate(sol_E):
    # 进一步用半角公式简化结果
    sol_half = sol.subs(sp.cos(phi), 2*sp.cos(phi/2)**2 - 1)
    simplified_sol = sp.simplify(sol_half)
    display(Math(rf"E_{i+1} = " + sp.latex(simplified_sol)))
```

### $W\not =0, \mu\not=0, q=0, \phi=\pi$

```python
mu = sp.symbols('mu', real=True)
# 区域 I (S): E=0, q=0, mu_S=mu
H_I_zero = (H1 - E * sp.eye(4)).subs({mu_S: mu, q: 0, E: 0})
det_I = sp.factor(H_I_zero.det())

display(Math(r"\det(\mathcal{H}_I)_{E=0} = " + sp.latex(det_I) + " = 0"))

# 区域 II (N): E=0, q=0, mu_N=mu
H_II_zero = (H2 - E * sp.eye(4)).subs({mu_N: mu, q: 0, E: 0})
det_II = sp.factor(H_II_zero.det())

display(Math(r"\det(\mathcal{H}_{II})_{E=0} = " + sp.latex(det_II) + " = 0"))

# 区域 III (S): E=0, q=0, mu_S=mu
H_III_zero = (H3 - E * sp.eye(4)).subs({mu_S: mu, q: 0, E: 0, phi: sp.pi})
det_III = sp.factor(H_III_zero.det())

display(Math(r"\det(\mathcal{H}_{III})_{E=0} = " + sp.latex(det_III) + " = 0"))
```

$$
k_y^{(I)}=\pm \frac{\mu}{v}, \kappa_y^{(I)}=-\frac{\Delta_0}{v}
$$
$$
k_{y}^{(II)}=\pm \frac{\mu}{v}, \kappa_y^{(II)}=0
$$
$$
k_y^{(III)}=\pm \frac{\mu}{v}, \kappa_y^{(III)}=\frac{\Delta_0}{v}
$$

```python
k_F = mu / v
kappa_S = Delta0 / v

def normalize_first_element(vec):
    """找到第一个非零分量并将其化为 1"""
    for i in range(len(vec)):
        if vec[i] != 0:
            # 提取该分量并化简，然后全向量除以它
            factor = vec[i]
            return sp.simplify(vec / factor)
    return vec

ky_branches = [k_F, -k_F]

for ky_val in ky_branches:
    display(Math(r"\blacktriangleright \text{正在计算全局动量分支: } k_y = " + sp.latex(ky_val)))
    
    # --- 区域 I (S): 向左衰减 ---
    M_I = H_I_zero.subs({kyI: ky_val, kappaI: -kappa_S})
    ns_I = [normalize_first_element(sp.simplify(vec)) for vec in M_I.nullspace()]
    display(Math(r"\text{区域 I (S) 归一化基底：}"))
    for v_i in ns_I: display(v_i)
        
    # --- 区域 II (N): 传播态 ---
    M_II = H_II_zero.subs({kyII: ky_val, kappaII: 0})
    ns_II = [normalize_first_element(sp.simplify(vec)) for vec in M_II.nullspace()]
    display(Math(r"\text{区域 II (N) 归一化基底：}"))
    for v_ii in ns_II: display(v_ii)

    # --- 区域 III (S): 向右衰减 ---
    M_III = H_III_zero.subs({kyIII: ky_val, kappaIII: kappa_S})
    ns_III = [normalize_first_element(sp.simplify(vec)) for vec in M_III.nullspace()]
    display(Math(r"\text{区域 III (S) 归一化基底：}"))
    for v_iii in ns_III: display(v_iii)
    
    print("\n" + "-"*60 + "\n")
```

$$\Psi_I(y) =\begin{bmatrix} 1 \\ i \\ i \\ -1 \end{bmatrix} e^{i \frac{\mu}{v} y} e^{\frac{\Delta_0}{v} (y+W/2)}$$
$$\Psi_{III}(y) = \begin{bmatrix} 1 \\ i \\ i \\ -1 \end{bmatrix} e^{i \frac{\mu}{v} y} e^{-\frac{\Delta_0}{v} (y-W/2)}$$

```python
import sympy
from sympy import I, Matrix, simplify, symbols
from sympy.physics.quantum import TensorProduct
from IPython.display import display, Math

def pauli_matrices():
    s0 = Matrix([[1, 0], [0, 1]])
    sx = Matrix([[0, 1], [1, 0]])
    sy = Matrix([[0, -I], [I, 0]])
    sz = Matrix([[1, 0], [0, -1]])
    return s0, sx, sy, sz

s0, sx, sy, sz = pauli_matrices()
t0, tx, ty, tz = s0, sx, sy, sz

zeta_p = Matrix([1,  I,  I, -1]) / 2
zeta_m = Matrix([1, -I, -I, -1]) / 2

zeta1 = simplify((zeta_p + zeta_m) / 2)
zeta2 = simplify((zeta_p - zeta_m) / (2 * I))

def bracket(bra, op, ket):
    return simplify(bra.H * op * ket)[0]

Op89 = TensorProduct(tz, sy)
Op1011 = TensorProduct(ty, s0)


display(Math(r'\text{Defined Basis Elements:}'))
display(Math(r'\zeta_1^T = ' + sympy.latex(zeta1.T)))
display(Math(r'\zeta_2^T = ' + sympy.latex(zeta2.T)))

print("-" * 30)

val8a = bracket(zeta1, Op89, zeta2)
val8b = bracket(zeta2, Op89, zeta1)
display(Math(r'\text{Eq (8): } \langle \zeta_1 | \tau_z \sigma_y | \zeta_2 \rangle = %s, \quad \langle \zeta_2 | \tau_z \sigma_y | \zeta_1 \rangle = %s' 
             % (sympy.latex(val8a), sympy.latex(val8b))))

val9a = bracket(zeta1, Op89, zeta1)
val9b = bracket(zeta2, Op89, zeta2)
display(Math(r'\text{Eq (9): } \langle \zeta_{1,2} | \tau_z \sigma_y | \zeta_{1,2} \rangle = %s, %s' 
             % (sympy.latex(val9a), sympy.latex(val9b))))

val10a = -bracket(zeta1, Op1011, zeta2)
val10b = bracket(zeta2, Op1011, zeta1)
display(Math(r'\text{Eq (10): } -\langle \zeta_1 | \tau_y \sigma_0 | \zeta_2 \rangle = %s, \quad \langle \zeta_2 | \tau_y \sigma_0 | \zeta_1 \rangle = %s' 
             % (sympy.latex(val10a), sympy.latex(val10b))))

val11a = bracket(zeta1, Op1011, zeta1)
val11b = bracket(zeta2, Op1011, zeta2)
display(Math(r'\text{Eq (11): } \langle \zeta_{1,2} | \tau_y \sigma_0 | \zeta_{1,2} \rangle = %s, %s' 
             % (sympy.latex(val11a), sympy.latex(val11b))))
```
